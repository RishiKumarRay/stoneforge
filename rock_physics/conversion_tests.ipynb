```python
import numpy as np
import matplotlib.pyplot as plt
import routinas_poco as uff

# convertion module
import log_unit_converter as uff2
```

```python
# Creating random logs
sz = 100

np.random.seed(seed=99)

DT = np.random.uniform(low=20.0, high=100.0, size=sz)
GR = np.random.uniform(low=0.0, high=150.0, size=sz)
ILD = np.random.uniform(low=0.0, high=5.0, size=sz)
NPHI = np.random.uniform(low=-0.15, high=0.4, size=sz)
RHOB = np.random.uniform(low=2.0, high=2.8, size=sz)
```

```python
# Convertion Test - velocity

VP = uff2.velocity_si(DT)


NDT = uff2.vagarosity_sl(VP)
plt.plot(VP,DT,'.')
plt.show()

plt.plot(NDT,DT,'.')
plt.show()
```

```python
# Convertion Test - density

density = uff2.density_si(RHOB)
N_bulk = uff2.density_sl(density)
plt.plot(RHOB,density,'.')
plt.show()

plt.plot(RHOB,N_bulk,'.')
plt.show()
```

```python
# Convertion Test - neutron porosity

neutron_100 = uff2.neutron_porosity_percentage(NPHI)
neutron_fr = uff2.neutron_porosity_fraction (neutron_100)
plt.plot(NPHI,neutron_100,'.')
plt.show()

plt.plot(NPHI,neutron_fr,'.')
plt.show()
```
